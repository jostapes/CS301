## Problem 1

```python
def get_percent_nans(df, column_name):
    col = df[column_name]
    null = col.isnull().sum()
    total = len(df)
    return  null/total*100
```

```python
for col in my_df.columns:
    print('The \"{}\" columnn has {:.2f}% empty values'.format(col,get_percent_nans(my_df, col)))
```

## Problem 6

```python
def get_continent(country):
    d= {'North America':['United States', 'Mexico','Canada'],
        'South America':['Brazil','Colombia'],
        'Asia':['Mainland China', 'Singapore', 'Thailand',
                'Japan', 'Hong Kong', 'South Korea', 'Malaysia',
                'Taiwan', 'Macau', 'Vietnam', 'United Arab Emirates',
                'India', 'Phillipines', 'Russia', 'Cambodia',
                'Nepal', 'Sri Lanka'],
        'Europe':['Germany','France','Italy','UK','Beligium','Finland','Spain','Sweden'],
        'Africa':['Ivory Coast', 'Oceania'],
        'Australia':['Australia'],
       }
    for key in d:
        if country in d[key]:
            return key
get_continent('UK')
```

## Problem 7

```python
my_df['Continent'] = my_df['Country/Region'].apply(get_continent)
my_df
```
